```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Sorcery_IOB2.csv')

# --- 1. Token-Level Distribution ---
# Extract the base entity type by stripping the 'B-' or 'I-' prefixes
df['Entity_Token_Class'] = df['IOB2'].apply(lambda x: x.split('-')[1] if '-' in x else x)
token_distribution = df['Entity_Token_Class'].value_counts()

print("=== Token-Level Distribution ===")
print(token_distribution)
print(f"Total Tokens: {len(df)}")

# --- 2. Entity-Mention Level Distribution ---
# Filter for tokens starting with 'B-' to count each unique entity phrase instance
entity_starts = df[df['IOB2'].str.startswith('B-', na=False)]
entity_distribution = entity_starts['IOB2'].apply(lambda x: x.split('-')[1]).value_counts()

# Count the 'O' (Outside) tokens separately since they represent non-entity tokens
o_count = (df['IOB2'] == 'O').sum()

print("\n=== Entity Mention Distribution (B- Tags) ===")
print(entity_distribution)
print(f"O (Non-entity tokens): {o_count}")

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('Planeswalker_IOB2.csv')

# --- 1. Token-Level Distribution ---
# Strip 'B-' and 'I-' prefixes to look at pure token classifications
df['Entity_Token'] = df['IOB2'].apply(lambda x: x.split('-')[-1] if '-' in x else x)
token_counts = df['Entity_Token'].value_counts()
token_pct = (token_counts / token_counts.sum()) * 100

# Sort descending for the bar chart
token_pct_ordered = token_pct.sort_values(ascending=False)

# --- 2. Entity-Mention Level Distribution ---
# Filter for tokens starting with 'B-' to isolate distinct entity occurrences
entity_starts = df[df['IOB2'].str.startswith('B-', na=False)]
entity_counts = entity_starts['IOB2'].apply(lambda x: x.split('-')[1]).value_counts()
entity_pct = (entity_counts / entity_counts.sum()) * 100
entity_pct_ordered = entity_pct.sort_values(ascending=False)

# Define a muted cool peach color palette
muted_cool_peach = ['#E8A598', '#F0B7A4', '#F7C9B1', '#FAD6C0', '#FCE3CE', '#FFF0DD', '#E29578', '#DDA15E']

# --- Plot 1: Token-Level Distribution ---
fig1, ax1 = plt.subplots(figsize=(10, 6))
sns.barplot(
    x=token_pct_ordered.values, 
    y=token_pct_ordered.index, 
    ax=ax1, 
    palette=sns.color_palette(muted_cool_peach[:len(token_pct_ordered)])
)
ax1.set_title('Distribution of Entity Tokens (Percentage)', fontsize=14, pad=15)
ax1.set_xlabel('Percentage (%)', fontsize=12)
ax1.set_ylabel('Entity Type', fontsize=12)

# Append percentage labels on the bars
for i, v in enumerate(token_pct_ordered.values):
    ax1.text(v + 0.5, i, f"{v:.2f}%", va='center', fontsize=10)

plt.tight_layout()
plt.savefig('Planeswalker_article_entity_token_distribution.png', dpi=300)
plt.close()

# --- Plot 2: Entity Mention Distribution ---
fig2, ax2 = plt.subplots(figsize=(10, 6))
sns.barplot(
    x=entity_pct_ordered.values, 
    y=entity_pct_ordered.index, 
    ax=ax2, 
    palette=sns.color_palette(muted_cool_peach[:len(entity_pct_ordered)])
)
ax2.set_title('Distribution of Distinct Entity Mentions (Excluding O, Percentage)', fontsize=14, pad=15)
ax2.set_xlabel('Percentage (%)', fontsize=12)
ax2.set_ylabel('Entity Type', fontsize=12)

# Append percentage labels on the bars
for i, v in enumerate(entity_pct_ordered.values):
    ax2.text(v + 0.5, i, f"{v:.2f}%", va='center', fontsize=10)

plt.tight_layout()
plt.savefig('Planeswalker_article_entity_mention_distribution.png', dpi=300)
plt.close()
```
