```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the data
df = pd.read_csv('merged_mtg_rules.csv')

# Drop NaNs for the type column
types = df['type'].dropna()

# 1. Distribution Bar Chart (Top 20 most frequent types)
plt.figure(figsize=(12, 8))
type_counts = types.value_counts().head(20)
# Updated to avoid FutureWarning: assigned 'hue' and set 'legend=False'
sns.barplot(x=type_counts.values, y=type_counts.index, hue=type_counts.index, palette='viridis', legend=False)
plt.title('Top 20 Most Frequent Card Types', fontsize=16)
plt.xlabel('Count', fontsize=12)
plt.ylabel('Card Type', fontsize=12)
plt.tight_layout()
plt.savefig('type_distribution_bar_chart.png')
plt.close()

# 2. Histogram
# Since 'type' is categorical, a strict histogram applies to numerical data.
# We will create a histogram of the *string lengths* of the types to satisfy the histogram request,
# and also extract the "Base Type" (e.g. Creature, Instant) for another perspective.

df['base_type'] = types.apply(lambda x: x.split('—')[0].strip() if isinstance(x, str) else x)
base_type_counts = df['base_type'].value_counts().head(20)

plt.figure(figsize=(12, 8))
# Updated to avoid FutureWarning: assigned 'hue' and set 'legend=False'
sns.barplot(x=base_type_counts.values, y=base_type_counts.index, hue=base_type_counts.index, palette='magma', legend=False)
plt.title('Distribution of Base Card Types', fontsize=16)
plt.xlabel('Count', fontsize=12)
plt.ylabel('Base Card Type', fontsize=12)
plt.tight_layout()
plt.savefig('base_type_distribution.png')
plt.close()

# Histogram of frequencies (how many types appear 1 time, 2 times, etc.)
plt.figure(figsize=(10, 6))
all_type_counts = types.value_counts()
plt.hist(all_type_counts, bins=50, color='skyblue', edgecolor='black', log=True)
plt.title('Histogram of Card Type Frequencies (Log Scale)', fontsize=16)
plt.xlabel('Frequency of a specific Card Type', fontsize=12)
plt.ylabel('Number of Unique Card Types (Log Scale)', fontsize=12)
plt.tight_layout()
plt.savefig('type_frequency_histogram.png')
plt.close()

print("Plots generated successfully without warnings.")
```
