```python
import pandas as pd
import csv

# Load the dataset
df = pd.read_csv('BasicArticle_IOB2.csv')

# Open a new file to write out the data using csv.writer to safely handle embedded commas
with open('BasicArticle_CW.csv', 'w', encoding='utf-8', newline='') as f:
    writer = csv.writer(f, lineterminator='\n')
    writer.writerow(['Token', 'IOB2'])  # Header
    
    for i in range(len(df)):
        token = str(df.loc[i, 'Token'])
        iob2 = str(df.loc[i, 'IOB2'])
        
        # Check if a word ends in punctuation (longer than 1 character and ends with '.', '?', or '!')
        if len(token) > 1 and token[-1] in ['.', '?', '!']:
            punctuation = token[-1]
            word = token[:-1]
            
            # Write the stripped word with its original IOB2 tag
            writer.writerow([word, iob2])
            
            # Place the punctuation on a new line with an 'O' tag in column two
            writer.writerow([punctuation, 'O'])
            
            # Apply sentence separation logic after the punctuation line
            if i + 1 < len(df):
                next_token = str(df.loc[i + 1, 'Token'])
                if next_token and next_token[0].isupper():
                    f.write('\n')  # Insert a blank line separating sentences
            f.write('\n')
                    
        else:
            # Regular flow for standard tokens or isolated punctuation
            writer.writerow([token, iob2])
            
            # Sentence separation logic for existing isolated punctuation
            if any(p in token for p in ['.', '?', '!']) and not (iob2.startswith('B') or iob2.startswith('I')) and i + 1 < len(df):
                next_token = str(df.loc[i + 1, 'Token'])
                if next_token and next_token[0].isupper():
                    f.write('\n')  # Insert a blank line separating sentences
```
